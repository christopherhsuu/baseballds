```python
import re

def rename_columns(df):
    # Define the mapping from numbers to new names
    mapping = {
        '1': 'Hip', '2': 'RHip', '3': 'RKnee', '4': 'RAnkle',
        '5': 'RBigToe', '6': 'RSmallToe', '7': 'RHeel', '8': 'LHip',
        '9': 'LKnee', '10': 'LAnkle', '11': 'LBigToe', '12': 'LSmallToe',
        '13': 'LHeel', '14': 'Neck', '15': 'Head', '16': 'Nose',
        '17': 'RShoulder', '18': 'RElbow', '19': 'RWrist', '20': 'LShoulder',
        '21': 'LElbow', '22': 'LWrist'
    }

    rename_dict = {}
    # Regular expression to find numbers at the end of the column names
    pattern = re.compile(r'(\d+)$')

    # Loop over column names and apply mapping based on found numbers
    for col in df.columns:
        match = pattern.search(col)
        if match:
            num = match.group(1)
            if num in mapping:
                new_name = col[:match.start()] + mapping[num]
                rename_dict[col] = new_name

    # Rename the columns in the dataframe
    df.rename(columns=rename_dict, inplace=True)

# Usage
rename_columns(p1)
print(p1.columns)
```
