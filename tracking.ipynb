```python
import re

def rename_columns(df):
    # Define the mapping from numbers to new names
    mapping = {
        '1': 'Hip', '2': 'RHip', '3': 'RKnee', '4': 'RAnkle',
        '5': 'RBigToe', '6': 'RSmallToe', '7': 'RHeel', '8': 'LHip',
        '9': 'LKnee', '10': 'LAnkle', '11': 'LBigToe', '12': 'LSmallToe',
        '13': 'LHeel', '14': 'Neck', '15': 'Head', '16': 'Nose',
        '17': 'RShoulder', '18': 'RElbow', '19': 'RWrist', '20': 'LShoulder',
        '21': 'LElbow', '22': 'LWrist'
    }

    rename_dict = {}
    # Regular expression to find numbers at the end of the column names
    pattern = re.compile(r'(\d+)$')

    # Loop over column names and apply mapping based on found numbers
    for col in df.columns:
        match = pattern.search(col)
        if match:
            num = match.group(1)
            if num in mapping:
                new_name = col[:match.start()] + mapping[num]
                rename_dict[col] = new_name

    # Rename the columns in the dataframe
    df.rename(columns=rename_dict, inplace=True)

# Usage
rename_columns(p1)
print(p1.columns)
```

```python
import os
import pandas as pd
import re

BASE_DIR = r"C:\Users\chris\Documents\GitHub\baseballds\vids"


def clean_dataframe(df):
    df.columns = df.iloc[0]
    df = df[1:].reset_index(drop=True)

    columns_remove = [col for i, col in enumerate(df.columns[4:], start=4) if (i - 4) % 3 == 0 and 'Z' in col]
    df.drop(columns=columns_remove, inplace=True)

    df.rename(columns={df.columns[0]: 'Frame#', df.columns[1]: 'Time'}, inplace=True)

    mapping = {
        '1': 'Hip', '2': 'RHip', '3': 'RKnee', '4': 'RAnkle',
        '5': 'RBigToe', '6': 'RSmallToe', '7': 'RHeel', '8': 'LHip',
        '9': 'LKnee', '10': 'LAnkle', '11': 'LBigToe', '12': 'LSmallToe',
        '13': 'LHeel', '14': 'Neck', '15': 'Head', '16': 'Nose',
        '17': 'RShoulder', '18': 'RElbow', '19': 'RWrist', '20': 'LShoulder',
        '21': 'LElbow', '22': 'LWrist'
    }

    rename_dict = {}
    pattern = re.compile(r'(\d+)$')

    for col in df.columns:
        match = pattern.search(col)
        if match:
            num = match.group(1)
            if num in mapping:
                new_name = col[:match.start()] + mapping[num]
                rename_dict[col] = new_name

    df.rename(columns=rename_dict, inplace=True)
    return df

def process_trc_files(base_dir):
    for i in range(100):
        pitch_folder = os.path.join(base_dir, f"pitch{i}_Sports2D")
        if not os.path.exists(pitch_folder):
            continue 


        for filename in os.listdir(pitch_folder):
            if filename.endswith(".trc") and "px_person00" in filename:
                file_path = os.path.join(pitch_folder, filename)
                print(f"Processing: {file_path}")

                df = pd.read_csv(file_path, sep='\t', skiprows=3) 
                df = clean_dataframe(df)

                cleaned_file_path = os.path.join(pitch_folder, f"cleaned_{filename}")
                df.to_csv(cleaned_file_path, sep='\t', index=False)
                print(f"Saved cleaned file: {cleaned_file_path}")
```
